### 2. The dataframe we will use is a scrap from Wikipedia

```python
# Install and import the libraries 

import urllib.request
import pandas as pd

# ! pip install bs4
from bs4 import BeautifulSoup

print ('Libraries imported')
```

```python
# Get the webpage where to extract the data

URL = 'https://en.wikipedia.org/wiki/Districts_of_Barcelona'
page = urllib.request.urlopen(URL)


# Scrap the data

soup = BeautifulSoup(page, "lxml")
all_tables=soup.find_all("table")
right_table=soup.find('table', class_='wikitable')


# Analyse the source code and classsify the data in columns

A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
H=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==8:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        nlnk=cells[5].findAll('a')
        F.append(nlnk[1].contents[0])
        G.append(cells[6].find(text=True))
        H.append(cells[7].find(text=True))
```

```python
# Transfrom the data into a readable dataframe with pandas

df=pd.DataFrame(A,columns=["Number"])
df["District"]=B
df["Size"]=C
df["Population"]=D
df["Density"]=E
df["Neighborhoods"]=F
df["Councilman"]=G
df["Party"]=H

df.head(15)
```
